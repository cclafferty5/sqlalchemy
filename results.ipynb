```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
x = [1000, 10000, 100000, 1000000]
data = [(0.0025856494903564453, 0.0011088848114013672, 1.0), (0.00926971435546875, 0.004860401153564453, 1.0), (0.07501626014709473, 0.042181968688964844, 1.0), (1.9151484966278076, 1.536468267440796, 0.18991899189918993)]
no_cache, cache, hit_rate = zip(*data)
print(no_cache)

axes = plt.plot(x, no_cache, label = "no_cache")
axes2 = plt.plot(x, cache, label = "cached")
plt.xscale("log")
plt.legend(["no_cache", "cached"])
plt.title("# of Rows vs query speed in seconds")
plt.xlabel = "# of Rows"
plt.ylabel = "seconds"
plt.savefig('result1.png')
plt.show()
```
